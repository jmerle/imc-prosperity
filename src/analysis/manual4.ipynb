```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
```

```python
available_capital = 750_000

def investment(percentage: float) -> float:
    return (percentage / 100) * available_capital

def fee(percentage: float) -> float:
    return 1.2 * available_capital * (percentage / 100) ** 2

def break_even_return(percentage: float) -> float:
    return fee(percentage) / investment(percentage) * 100
```

```python
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(111)
ax2 = ax1.twinx()

x = np.linspace(0, 100, 101)
ax1.plot(x, investment(x), label="Investment = (percentage / 100) * 750,000")
ax1.plot(x, fee(x), label="Fee = 1.2 * 750,000 * (percentage / 100) ^ 2")
ax2.plot(x, break_even_return(x), label="Break-even return = fee / investment * 100", color="g")

ax1.set_xlabel("Percentage")
ax1.set_ylabel("Seashells")
ax2.set_ylabel("Percentage")

formatter = FuncFormatter(lambda x, p: f"{x:,.0f}")
ax1.get_xaxis().set_major_formatter(formatter)
ax1.get_xaxis().set_minor_formatter(formatter)
ax1.get_yaxis().set_major_formatter(formatter)
ax1.get_yaxis().set_minor_formatter(formatter)

intersect_x = 250 / 3
intersect_y = investment(intersect_x)
ax1.plot(intersect_x, intersect_y, "ro", label=f"Intersection ({intersect_x:,.2f}, {intersect_y:,.0f})")

ax1.set_xticks(np.arange(0, 101, 10))
ax1.set_yticks(np.arange(0, 1_000_001, 100_000))

ax1.grid()
fig.legend(loc="upper left", bbox_to_anchor=(0, 1), bbox_transform=ax1.transAxes);
```
